# 4. Explore trends and patterns using data visualization

```python
plt.figure(figsize=(12,8))
sns.barplot(x='home_team', y='runs', data=Data_worldcup)
plt.title('Runs Scored by Home Team')
plt.xlabel('Team')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='away_team', y='runs', data=Data_worldcup)
plt.title('Runs Scored by Away Team')
plt.xlabel('Team')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(24, 16))
sns.barplot(x='batsman1_name', y='batsman1_runs', data=Data_worldcup, estimator=sum, ci=None)
plt.title('Total Runs Scored by Batsmen')
plt.xlabel('Batsman Name')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# As we can not clearly see what is the output of this graph, so I need to adjust it for only top batsman
```

```python
top_batsmen = Data_worldcup.groupby('batsman1_name')['batsman1_runs'].sum().nlargest(10)  # Adjust the number of top batsmen as needed
plt.figure(figsize=(12, 8))
sns.barplot(x=top_batsmen.index, y=top_batsmen.values)
plt.title('Total Runs Scored by Top Batsmen')
plt.xlabel('Batsman Name')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# 5. Checking for duplicate values

```python
duplicated_values_count=len(Data_worldcup[Data_worldcup.duplicated()])
print("Number of duplicate values:",duplicated_values_count)
```

```python
duplicates = Data_worldcup.duplicated()
print("Duplicate Rows:")
print(Data_worldcup[duplicates])
```

```python
Data_worldcup= Data_worldcup.drop_duplicates()
Data_worldcup.duplicated().sum()
```

# 6. Checking for missing values

```python
missing_values = Data_worldcup.isnull().sum()
print("Missing Values:")
print(missing_values)
```

# 7. Visualizing Missing Values

```python
plt.figure(figsize=(10, 6))
sns.heatmap(Data_worldcup.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

```python
# Handling missing value by "Drop Missing Values": Remove rows or columns with missing values using dropna().
```

```python
Data_worldcup = Data_worldcup.dropna()
```

```python
# Verify Missing Values Handling: Check if missing values have been handled as intended.
```

```python
missing_values_after_handling = Data_worldcup.isnull().sum()
print("Missing Values after Handling:")
print(missing_values_after_handling)
```
